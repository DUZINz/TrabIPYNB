```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Salary_Data.csv')
df.head()  # Exibe as primeiras linhas do dataframe
```

```python
salaries = df['Salary'].to_numpy()
years_experience = df['YearsExperience'].to_numpy()
```

```python
mean_salary = np.mean(salaries)
print(f"Média de salarios: {mean_salary:.2f} anos")
```

```python
salary_above_5_years = salaries[years_experience > 5]
mean_salary_above_5_years = np.mean(salary_above_5_years)
print(f"Salário médio de funcionários com mais de 5 anos de serviço: {mean_salary_above_5_years:.2f}")
```

```python
plt.figure(figsize=(10, 6))
plt.bar(years_experience, salaries, color='green')
plt.xlabel('Anos de Experiência')
plt.ylabel('Salário')
plt.title('Salário por Anos de Experiência')
plt.show()
```

```python
increased_salaries = salaries * 1.10
df['Increased_Salary'] = increased_salaries  # Adicionando ao dataframe original para visualização

print("Salários após aumento de 10%:")
print(df[['YearsExperience', 'Salary', 'Increased_Salary']].head())
```
