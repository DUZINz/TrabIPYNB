```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Netflix TV Shows and Movies.csv')
df.head()
```

```python
df.head(10)
```

```python
df.tail(10)
```

```python
# Maior imdb_score
max_score_movie = df.loc[df['imdb_score'].idxmax()]
print("Filme com maior imdb_score:")
print(max_score_movie[['title', 'imdb_score']])

# Menor imdb_score
min_score_movie = df.loc[df['imdb_score'].idxmin()]
print("\nFilme com pior imdb_score:")
print(min_score_movie[['title', 'imdb_score']])
```

```python
# Convertendo runtime para numérico (removendo possíveis caracteres indesejados)
df['runtime'] = pd.to_numeric(df['runtime'], errors='coerce')

# Média de duração dos shows de TV
tv_shows_mean_runtime = df[df['type'] == 'SHOW']['runtime'].mean()
print(f"Média de duração dos shows de TV: {tv_shows_mean_runtime:.2f} minutos")

# Média de duração dos filmes
movies_mean_runtime = df[df['type'] == 'MOVIE']['runtime'].mean()
print(f"Média de duração dos filmes: {movies_mean_runtime:.2f} minutos")
```

```python
# Show de TV com maior duração
longest_tv_show = df[df['type'] == 'SHOW'].loc[df[df['type'] == 'SHOW']['runtime'].idxmax()]
print("Show de TV com maior duração:")
print(longest_tv_show[['title', 'runtime']])

# Filme com maior duração
longest_movie = df[df['type'] == 'MOVIE'].loc[df[df['type'] == 'MOVIE']['runtime'].idxmax()]
print("\nFilme com maior duração:")
print(longest_movie[['title', 'runtime']])
```

```python
movies_per_year = df[df['type'] == 'MOVIE'].groupby('release_year').size()
most_movies_year = movies_per_year.idxmax()
print(f"Ano com maior quantidade de filmes avaliados: {most_movies_year}")
```

```python
age_certification_counts = df['age_certification'].value_counts()
print("Categorias de age_certification e quantidade de filmes:")
print(age_certification_counts)
```

```python
missing_data = df.isnull().sum()
print("Colunas com campos vazios e quantidade:")
print(missing_data[missing_data > 0])
```

```python
# Média de scores para filmes por ano
movies_scores = df[df['type'] == 'MOVIE'].groupby('release_year')['imdb_score'].mean()

plt.figure(figsize=(10, 5))
plt.plot(movies_scores, label='Movies')
plt.xlabel('Ano')
plt.ylabel('Média IMDb Score')
plt.title('Média de IMDb Score por Ano - Filmes')
plt.legend()
plt.show()

# Média de scores para shows por ano
shows_scores = df[df['type'] == 'SHOW'].groupby('release_year')['imdb_score'].mean()

plt.figure(figsize=(10, 5))
plt.plot(shows_scores, label='Shows', color='red')
plt.xlabel('Ano')
plt.ylabel('Média IMDb Score')
plt.title('Média de IMDb Score por Ano - Shows')
plt.legend()
plt.show()
```

```python
# Total de votos para filmes e shows
movie_votes = df[df['type'] == 'MOVIE']['imdb_votes'].sum()
show_votes = df[df['type'] == 'SHOW']['imdb_votes'].sum()

# Criando gráfico de proporção
labels = ['Movies', 'Shows']
votes = [movie_votes, show_votes]

plt.figure(figsize=(7, 7))
plt.pie(votes, labels=labels, colors= 'green', autopct='%1.1f%%', startangle=140)
plt.title('Proporção de Votos IMDb - Shows vs Filmes')
plt.axis('equal')  # Para deixar o gráfico circular
plt.show()
```
